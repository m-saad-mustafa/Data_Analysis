```python
# Importation of Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Read CSV file 
df = pd.read_csv(r'C:\Users\M Saad Mustafa\OneDrive\Desktop\Numpy prac\stdscores data\std_scores.csv')
#Drop Particular Columns in data
df = df.drop('part_time_job',axis=1)
df = df.drop('extracurricular_activities',axis=1)
#To check details of data 
print(df.head())
##check missing values 
print(df.isnull().sum())



df.info()
```

Gender Discrimination


```python
plt.figure(figsize=(3,4.5))
plt.title('Gender Discrimination')
ax =sns.countplot(data=df, x='gender')
ax.bar_label(ax.containers[0])
plt.show()
#Through the result of this data we analyze that no. of males are more than no. of females.
```

Weekly Study Impact on Student Marks


```python
gb = df.groupby('weekly_self_study_hours').agg({'math_score':'mean','computer_score':'mean','physics_score':'mean'})
# print(gb)
plt.figure(figsize=(15,15))
sns.heatmap(gb, annot=True)
plt.title('Impact of Weekly Study on Subject Marks')
plt.show
# Through this we analyze that  weekly study hours have negligible impact on scores.
```

CountPlot of Career_Aspiration


```python
plt.figure(figsize=(33,13))
bx = sns.countplot(data=df,x='career_aspiration')
bx.bar_label(bx.containers[0])
plt.show()
#Through the output of this data we can analyze that maximum students want to become Software Engineer and only fews wants to become waiter.
```
